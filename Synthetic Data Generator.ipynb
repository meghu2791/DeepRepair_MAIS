## Introduce errors in the dataset into df_fourth dataframe on the selected columns.

```python

columns = df_first.shape[1]
rowsNum = df_first.shape[0]

def modifyValue(value):
    wordList = list(value)
    
    randomPosition = random.randint(0, len(value) - 1)
    currentChar = wordList[randomPosition]
    
    lowercase = set(string.ascii_lowercase)
    uppercase = set(string.ascii_uppercase)
    digits = set(string.digits)
    
    while currentChar.isspace() is True:
        if len(value):
            randomPosition = random.randint(0, len(value) - 1)
            currentChar = wordList[randomPosition]
    
    if currentChar.isalpha():
        if currentChar.islower():
            char = random.choice(list(lowercase - set(currentChar)))
        else:
            char = random.choice(list(uppercase - set(currentChar)))
    else:
        char = random.choice(list(digits - set(currentChar)))
    
    #wordList[randomPosition] = str(char)
    if wordList:
        wordList[randomPosition] = str('x')
    return ''.join(wordList)

for row in range(sample_size):
        random_row = random.randint(0,rowsNum-1)
        random_cell_1 = random.randint(0,columns-2)
        #print(random_cell_1)
        #if col_list[random_cell_1] == 'sex' or col_list[random_cell_1] == 'race':
        #    continue
        if random_row in tempList:
            row -= 1
        else:
            value = df_first.iloc[random_row,random_cell_1]
            if value == float(np.nan):
                df_first.iloc[random_row,random_cell_1] = str('nan')
            if value or value!=float(np.nan) or value!='nan':
                df_first.iloc[random_row,random_cell_1] = modifyValue(value)
                tempList.append(random_row)
                df_first.iloc[random_row, -1] = 0
    #df_fourth.at[index, 'label'] = 0
```
